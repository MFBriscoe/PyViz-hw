# San Francisco Rental Prices Dashboard

In this notebook, you will compile the visualizations from the previous analysis into functions that can be used for a Panel dashboard.

```python
# imports
import panel as pn
pn.extension('plotly')
import plotly.express as px
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
import os
from pathlib import Path
from dotenv import load_dotenv
```

```python
# Read the Mapbox API key
load_dotenv()
mapbox_token = os.getenv("mapbox")
```

# Import Data

```python
# Import the CSVs to Pandas DataFrames
file_path = Path("Data/sfo_neighborhoods_census_data.csv")
sfo_data = pd.read_csv(file_path, index_col="year")

file_path = Path("Data/neighborhoods_coordinates.csv")
df_neighborhood_locations = pd.read_csv(file_path)
```

```python
# Define Panel Visualization Functions
def housing_units_per_year():
    """Housing Units Per Year."""
    
    # YOUR CODE HERE!
df_housing_units = sfo_data["housing_units"].groupby(sfo_data.index).mean()
df_housing_units

fig_housing_units = plt.figure()
plot_housing_units = df_housing_units.plot.bar()
plot_housing_units.set_xlabel("Year", fontsize=14)
plot_housing_units.set_ylabel("Housing Units", fontsize=14)
plot_housing_units.set_title("Housing Units in San Francisco from 2010 to 2016", fontsize=16)
plot_housing_units.set_ylim(ymin=370000, ymax=385000);
```

```python
def average_gross_rent():
    """Average Gross Rent in San Francisco Per Year."""
    
    # YOUR CODE HERE!
avg_rent_sale_data = sfo_data[['sale_price_sqr_foot','gross_rent']].groupby('year').mean()
avg_rent_sale_data

avg_gross_rent = avg_rent_sale_data[['gross_rent']]
agr=avg_gross_rent.plot(title='Average Gross Rent in San Francisco');
```

```python
def average_sales_price():
    """Average Sales Price Per Year."""
    
    # YOUR CODE HERE!
avg_sale_price = avg_rent_sale_data[['sale_price_sqr_foot']]
asp = avg_sale_price.plot(title='Average Sale Price Per Square Foot in San Francisco');
```

```python
def average_price_by_neighborhood():
    """Average Prices by Neighborhood."""
    
    # YOUR CODE HERE!
new_housing_df = sfo_data.groupby(['year','neighborhood']).mean()
new_housing_df.reset_index(inplace=True)
new_housing_df.head()

apn = new_housing_df.hvplot.line('year',
                           'sale_price_sqr_foot',
                           xlabel='Year',
                           ylabel='Average Sale Price/SF',
                           groupby='neighborhood'
)
```

```python
def top_most_expensive_neighborhoods():
    """Top 10 Most Expensive Neighborhoods."""
    
    # YOUR CODE HERE!
top_ten = new_housing_df.groupby(['neighborhood']).mean()
top_ten_df=top_ten.sort_values(by='sale_price_sqr_foot', ascending=False).head(10)
top_ten_df.reset_index(inplace=True)
top_ten_df

ballin = top_ten_df.hvplot.bar(
    x='neighborhood',
    y='sale_price_sqr_foot',
    xlabel='Neighborhood',
    ylabel='Average Sale Price / Sq. Ft.',
    title='Top 10 Most Expensive Neighborhoods',
    rot=70
)
```

```python
def parallel_coordinates():
    """Parallel Coordinates Plot."""
    
    # YOUR CODE HERE!
top_ten_df=top_ten_df.drop(columns='year')
top_ten_df.dropna()
top_ten_df

coord_plot=px.parallel_coordinates(top_ten_df, color='sale_price_sqr_foot')
```

```python
def parallel_categories():
    """Parallel Categories Plot."""
    
    # YOUR CODE HERE!
cat_plot = px.parallel_categories(
    data_frame=top_ten_df,
    dimensions=['neighborhood','sale_price_sqr_foot','housing_units','gross_rent'],
    color='sale_price_sqr_foot',
)
```

```python
def neighborhood_map():
    """Neighborhood Map"""
    
    # YOUR CODE HERE!
neighborhood_means = new_housing_df
neighborhood_means.head()

combined_df = pd.concat([df_neighborhood_locations, neighborhood_means], axis=1, join='inner')
combined_df=combined_df.drop(columns='neighborhood')
combined_df.head()

px.set_mapbox_access_token(mapbox_token)

map_1 = px.scatter_mapbox(
    combined_df,
    lat="Lat",
    lon="Lon",
    color="Neighborhood",
    size="sale_price_sqr_foot"
)
```

## Panel Dashboard

In this section, you will combine all of the plots into a single dashboard view using Panel. Be creative with your dashboard design!

```python
# YOUR CODE HERE!
#Create Rows
#housing averages
row_of_housing_stats = pn.Row(plot_housing_units, agr, asp, apn, ballin)

#trends
row_of_parallels = pn.Row(coord_plot, cat_plot)

#maps
row_of_maps = pn.Row(map_1)


plots_as_column = pn.Column(
    "#San Francisco Rental Property Dashboard", 
    "##General Plots",
    row_of_housing_stats,
    "##Statistical Trends",
    row_of_parallels,
    "##Maps",
    row_of_maps
)

sfo_dashboard = pn.Tabs(
    ("All Plots", plots_as_column),
    ("General Plots", row_of_housing_stats),
    ("Statistical Trends", row_of_parallels),
    ("Maps", row_of_maps)
)
sfo_dashboard
```

## Serve the Panel Dashboard

```python
sfo_dashboard.servable()
```
